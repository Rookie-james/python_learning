```python
import pandas as pd
import numpy as np
data06 = pd.read_csv(r"D:\ProgramData\python数据分析与挖掘\Python数据分析与挖掘【326758】了解一下数据的基本情况（一）\第9章 pandas模块的介绍\diamonds.csv")
data06.head()
```

```python
pd.pivot_table(data06, index='color',values='price',aggfunc='mean')
```

```python
pd.pivot_table(data06, index='color',columns='clarity',values='price',aggfunc='size')
```

```python
# 通过groupby方法，指定分组变量
grouped = data06.grouped(by=['colors','cut'])
# 对分组变量进行汇总
result = grouped.aggregate({'colors':np.size,
                            'carat':np.min,
                            'price':np.mean,
                            'face_width':np.max})
result
```
